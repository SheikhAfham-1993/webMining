```python
import pandas as pd
import spacy
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import classification_report, accuracy_score
from nltk.tokenize import RegexpTokenizer
import numpy as np
from nltk.corpus import stopwords
import nltk
from sklearn.model_selection import cross_val_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
```

```python
df = pd.read_csv('../Datasets/train/steam_ds.csv', encoding = 'ISO-8859-1')
```

```python
df.head()
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
df['user_review'] = df['user_review'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df.head()
```

```python
df['user_review'] = df['user_review'].str.replace('[^\w\s]',' ')
df.head()
```

```python
def space(comment):
    doc = nlp(comment)
    return " ".join([token.lemma_ for token in doc])
df['user_review']= df['user_review'].apply(space)
df.head()
```

```python
nltk.download('stopwords')
```

```python
stop = stopwords.words('english')
df['user_review'] = df['user_review'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df.head()
```

```python
labels = df[['user_suggestion']]
text = df[['user_review']]
```

```python
X_train, x_test, Y_train,Y_test  = train_test_split(text, labels, test_size = 0.15, random_state=42)
```

```python
def vectorize(data,tfidf_vect_fit):
    X_tfidf = tfidf_vect_fit.transform(data)
    words = tfidf_vect_fit.get_feature_names()
    X_tfidf_df = pd.DataFrame(X_tfidf.toarray())
    X_tfidf_df.columns = words
    return(X_tfidf_df)
```

```python
#token = RegexpTokenizer(r'[a-zA-Z0-9]+')
#cv = CountVectorizer(stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)

#vectorizer = TfidfVectorizer()
#X_train = vectorizer.fit_transform(X_train)

tfidf_vect = TfidfVectorizer()
tfidf_vect_fit=tfidf_vect.fit(X_train['user_review'])
X_train=vectorize(X_train['user_review'],tfidf_vect_fit)
```

```python
X_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
lr = LogisticRegression()
grid = {"C":np.logspace(-3,3,7), "penalty":["l1", "l2"]}
grid_lr = GridSearchCV(lr, grid, cv=5)
grid_lr.fit(X_train, Y_train.values.ravel())
```

```python
grid_lr.best_params_
```

```python
lr = LogisticRegression(C =1.0, penalty='l2')
```

```python
lr.fit(X_train, Y_train.values.ravel())
```

```python
predicted = lr.predict(x_test)
```

```python
print(accuracy_score(predicted, Y_test))
```

```python
lr_cv = LogisticRegression()
scores = cross_val_score(lr_cv,X_train,Y_train,cv=10)
print(scores)
scores.mean()
```

```python
lr_cv.fit(X_train, Y_train)
predicted_cv = lr_cv.predict(x_test)
print(accuracy_score(predicted_cv, Y_test))
```
