## Load the libraries

```python
import numpy as np
import tensorflow as tf
import pandas as pd
import spacy
from matplotlib import pyplot
from sklearn.metrics import accuracy_score,precision_score,recall_score
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense, Dropout
from tensorflow.keras.layers import SpatialDropout1D
from tensorflow.keras.layers import Embedding
from keras import layers
```

## Initialize variables

```python
oov_tok = '<OOV>'
```

## Initialize Methods

```python
def lemma(text):
    doc=nlp(text)
    return " ".join([token.lemma_ for token in doc])

def stopword(text):
    doc=nlp(text)
    return " ".join([token.lemma_ for token in doc if not token.is_stop])
```

## Read Dataset

```python
df = pd.read_csv('../../Datasets/train/steam_ds_preprocessed.csv')
```

```python
df = df[df.user_review.isna() == False]
```

## Preprocessing

```python
nlp = spacy.load('en_core_web_lg', disable=['parser', 'ner'])
```

```python
df['user_review'] = df['user_review'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
df['user_review'] = df['user_review'].str.replace('[^\w\s]','')
```

```python
df['user_review'] = df['user_review'].apply(lemma)
```

```python
df['user_review'] = df['user_review'].apply(stopword)
```

```python
document = df[['user_review']]
sentiment = df[['user_suggestion']]
```

## Data split

```python
X_train, X_test, y_train, y_test = train_test_split(document,  sentiment, test_size=0.20, random_state=42)
```

## Textual Data processing (Embeddings and Factorizing)

```python
sentiment_label = y_train.user_suggestion.factorize()
sentiment_label_test = y_test.user_suggestion.factorize()
```

```python
review = X_train.user_review.values
tokenizer = Tokenizer(num_words=5000, oov_token=oov_tok)
tokenizer.fit_on_texts(review)
vocab_size = len(tokenizer.word_index) + 1
encoded_docs = tokenizer.texts_to_sequences(review)
padded_sequence = pad_sequences(encoded_docs, maxlen=200)
```

```python
# Time based decay learning rate
initial_learning_rate = 0.001
epochs = 100
decay = initial_learning_rate / epochs
def lr_time_based_decay(epoch, lr):
    return lr * 1 / (1 + decay * epoch)
```

```python
callback = tf.keras.callbacks.LearningRateScheduler(lr_time_based_decay)
```

```python
adam_opt = tf.keras.optimizers.Adam(
    learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-07, amsgrad=False,
    name='Adam')
```

## Model Initialization

```python
embedding_vector_length = 32
model = Sequential(
[
    # Add an Embedding layer expecting input vocab of size 5000, and output embedding dimension of size 64 we set at the top
    tf.keras.layers.Embedding(5000, 64),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
#    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
    # use ReLU in place of tanh function since they are very good alternatives of each other.
    tf.keras.layers.Dense(64, activation='relu'),
    # Add a Dense layer with 6 units and softmax activation.
    # When we have multiple outputs, softmax convert outputs layers into a probability distribution.
    tf.keras.layers.Dense(6, activation='softmax')
]
)
#model.add(Embedding(vocab_size, embedding_vector_length, input_length=200) )
#model.add(SpatialDropout1D(0.25))
#model.add(LSTM(50, dropout=0.5, recurrent_dropout=0.5))
#model.add(Dropout(0.2))
#model.add(Dense(1, activation='sigmoid'))
#model.compile(loss='binary_crossentropy',optimizer=adam_opt, metrics=['accuracy'])
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

## Training Model

```python
history = model.fit(padded_sequence,sentiment_label[0],
                  validation_split=0.2, epochs=10, batch_size=32, callbacks=[callback])
```

```python
# plot loss during training
#pyplot.subplot(211)
fig=pyplot.figure(figsize=(15, 8))
pyplot.title('Loss')
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
fig.savefig("Loss.pdf")
pyplot.show()
```

```python
# plot accuracy during training
#pyplot.subplot(212)
fig_=pyplot.figure(figsize=(15, 8))
pyplot.title('Accuracy')
pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
fig_.savefig("Accuracy.pdf")
pyplot.show()
```

### Emotions

```python
df_emo = pd.read_csv('../../Datasets/train/text_emotion_updated.csv')
```

```python
df_emo.head()
```

```python
nlp = spacy.load('en_core_web_lg', disable=['parser', 'ner'])
```

```python
df_emo['text'] = df_emo['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
df_emo['text'] = df_emo['text'].str.replace('[^\w\s]','')
```

```python
df_emo['text'] = df_emo['text'].apply(lemma)
```

```python
df_emo['text'] = df_emo['text'].apply(stopword)
```

```python
document = df_emo[['text']]
emotions = df_emo[['emotions']]
```

```python
emotions['emotions']
```

```python
X_train, X_test, y_train, y_test = train_test_split(document,  emotions, test_size=0.10, random_state=42)
```

```python
label_tokenizer = Tokenizer()
label_tokenizer.fit_on_texts(emotions['emotions'])

emotions_label = np.array(label_tokenizer.texts_to_sequences(y_train['emotions']))#y_train.emotions.factorize()
emotions_label_test = np.array(label_tokenizer.texts_to_sequences(y_train['emotions']))#y_test.emotions.factorize()
```

```python
label_tokenizer.word_index
```

```python
review = X_train.text.values
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(review)
vocab_size = len(tokenizer.word_index) + 1
encoded_docs = tokenizer.texts_to_sequences(review)
padded_sequence = pad_sequences(encoded_docs, maxlen=200)
```

```python
# Time based decay learning rate
initial_learning_rate = 0.001
epochs = 100
decay = initial_learning_rate / epochs
def lr_time_based_decay(epoch, lr):
    return lr * 1 / (1 + decay * epoch)
```

```python
callback = tf.keras.callbacks.LearningRateScheduler(lr_time_based_decay)
```

```python
adam_opt = tf.keras.optimizers.Adam(
    learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-07, amsgrad=False,
    name='Adam')
opt = tf.keras.optimizers.Adam(lr=0.001, decay=1e-6)
```

```python
embedding_vector_length = 32
model = Sequential()
model.add(Embedding(vocab_size, 64))
model.add(Dropout(0.5))
model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)))
model.add(Dense(8, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy',optimizer=opt, metrics=['accuracy'])
#model.add(Embedding(vocab_size, embedding_vector_length, input_length=200) )
#model.add(SpatialDropout1D(0.25))
#model.add(LSTM(50, dropout=0.5, recurrent_dropout=0.5))
#model.add(Dropout(0.2))
#model.add(Dense(1, activation='sigmoid'))
#model.compile(loss='categorical_crossentropy',optimizer=adam_opt, metrics=['accuracy'])
#model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python

history = model.fit(padded_sequence, emotions_label, epochs=10, validation_split=0.2, batch_size=100, callbacks=[callback])

#history = model.fit(padded_sequence,emotions_label[0],
#                  validation_split=0.2, epochs=100, batch_size=100, callbacks=[callback])
```

```python
# plot loss during training
#pyplot.subplot(211)
fig=pyplot.figure(figsize=(15, 8))
pyplot.title('Loss')
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
fig.savefig("Loss.pdf")
pyplot.show()
```

```python
# plot accuracy during training
#pyplot.subplot(212)
fig_=pyplot.figure(figsize=(15, 8))
pyplot.title('Accuracy')
pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
fig_.savefig("Accuracy.pdf")
pyplot.show()
```
