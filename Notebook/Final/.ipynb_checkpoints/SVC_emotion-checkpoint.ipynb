```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import re
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from wordcloud import WordCloud, STOPWORDS
import spacy
from sklearn import preprocessing
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
import nltk
nltk.download('stopwords')
```

```python
df = pd.read_csv('../Datasets/train/text_emotion.csv')
```

```python
df.head()
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
df['text'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df.head()
```

```python
df['text'] = df['text'].str.replace('[^\w\s]','')
df.head()
```

```python
stop = stopwords.words('english')
df['text'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df.head()
```

```python
def space(comment):
    doc = nlp(comment)
    return " ".join([token.lemma_ for token in doc])
df['text']= df['text'].apply(space)
df.head()
```

```python
le = preprocessing.LabelEncoder()
df['emotions'] = le.fit_transform(df['emotions'])
df.head()
```

```python
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
text_counts = cv.fit_transform(df['text'])
X_train, X_test, Y_train, Y_test = train_test_split(text_counts,df['emotions'] ,test_size=0.25, random_state=5)
```

```python
from sklearn.svm import SVC
```

```python
svclassifier = SVC(kernel='sigmoid')
svclassifier.fit(X_train, Y_train)
#gnb.fit(data_train, target_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score
```

```python
predicted = svclassifier.predict(X_test)
accuracy_score = metrics.accuracy_score(predicted, Y_test)
print(str('{:04.2f}'.format(accuracy_score*100))+'%')
```

```python
recall = recall_score( Y_test, predicted, average='weighted')
print('Recall: %.3f' % recall)
```
