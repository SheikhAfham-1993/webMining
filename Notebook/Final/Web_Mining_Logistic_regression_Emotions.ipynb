```python
import pandas as pd
import spacy
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import classification_report, accuracy_score
from nltk.tokenize import RegexpTokenizer
import numpy as np
from nltk.corpus import stopwords
import nltk
from sklearn.model_selection import cross_val_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('../Datasets/train/text_emotion_updated.csv')
```

```python
df.head()
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
document = df['text']
sentiment = df['emotions']
```

```python
def space(comment):
    doc = nlp(comment)
    return " ".join([token.lemma_ for token in doc])
document = document.apply(space)
```

```python
document = document.apply(lambda x: " ".join(x.lower() for x in x.split()))
df.head()
```

```python
document = document.str.replace('[^\w\s]','')
df.head()
```

```python
def stopword(text):
    doc=nlp(text)
    return " ".join([token.lemma_ for token in doc if not token.is_stop])
```

```python
document = document.apply(stopword)
```

```python
X_train, X_test, y_train, y_test = train_test_split(document, sentiment, test_size=0.20, random_state=42)
```

```python
vectorizer = TfidfVectorizer(sublinear_tf=True, max_df=0.5, stop_words='english')
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
lr = LogisticRegression()
grid = {"C":np.logspace(-3,3,7), "penalty":["l1", "l2"]}
grid_lr = GridSearchCV(lr, grid, cv=10)
grid_lr.fit(X_train, y_train)
```

```python
grid_lr.best_params_
```

```python
lr = LogisticRegression(C =1.0, penalty='l2')
```

```python
lr.fit(X_train, y_train)
```

```python
predicted = lr.predict(X_test)
```

```python
print(accuracy_score(predicted, y_test))
```

```python
lr_cv = LogisticRegression()
scores = cross_val_score(lr_cv,X_train,y_train,cv=10)
print(scores)
scores.mean()
```

```python
lr_cv.fit(X_train, y_train)
predicted_cv = lr_cv.predict(X_test)
print(accuracy_score(predicted_cv, y_test))
```

```python
print(recall_score(predicted_cv, y_test, average='weighted'))
```
