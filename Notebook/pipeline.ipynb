```python
import numpy as np 
import pandas as pd 
import re
import nltk 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_spli
```

```python
df = pd.read_csv("../Datasets/train/steam_ds.csv")
```

```python
df.head()
```

```python
X = df['user_review']
y = df['user_suggestion']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
cv = CountVectorizer(stop_words='english',ngram_range = (1,1))
text_counts = cv.fit_transform(X_train)
label = cv.transform(X_test)
```

```python
pipe = Pipeline([('svc', SVC())])
```

```python
pipe.fit(text_counts, y_train)
```

```python
Pipeline(steps=[('svc', SVC())])
```

```python
pipe.score(text_counts, y_train)
```
